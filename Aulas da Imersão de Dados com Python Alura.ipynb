```python
df = df.rename(columns={
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
})

print("Colunas do DataFrame traduzidas:")
print(df.columns)
```

```python
df['senioridade'] = df['senioridade'].replace({
    'EN': 'Júnior',
    'MI': 'Pleno',
    'SE': 'Sênior',
    'EX': 'Executivo'
})

print("Categorias da coluna 'senioridade' traduzidas:")
print(df['senioridade'].value_counts())
```

```python
df['senioridade'].value_counts()
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
})

print("Categorias da coluna 'contrato' traduzidas:")
print(df['contrato'].value_counts())
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'Média',
    'L': 'Grande',
    'S': 'Pequena'
})

print("Categorias da coluna 'tamanho_empresa' traduzidas:")
print(df['tamanho_empresa'].value_counts())
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Totalmente Remoto'
})

print("Categorias da coluna 'remoto' traduzidas:")
print(df['remoto'].value_counts())
```

```python
import numpy as np

#Criação de um dataframe de teste para usar de exemplo
df_salarios = pd.DataFrame({
    'Nome':['Tiago','Gabriel', 'Cristiano', 'Warlley',"Felipe"],
    'Salario':[5000, np.nan, 4000, np.nan, 100000]
})
#calcula a média salarial e substitui os nulos pela média e arredonda os valores
df_salarios['salario_media'] = df_salarios['Salario'].fillna(df_salarios['Salario'].mean().round(2))

#calcula a mediana salarial e substitui os nulos pela mediana
df_salarios['salario_mediana'] = df_salarios['Salario'].fillna(df_salarios['Salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'Dia': ['Seg', 'Ter', 'Qua', 'Qui', 'Sex'],
    'Temperatura': [30, np.nan, np.nan, 28, 27],

})

df_temperaturas['Preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    'Dia': ['Seg', 'Ter', 'Qua', 'Qui', 'Sex'],
    'Temperatura': [30, np.nan, np.nan, 28, 27],

})

df_temperaturas['Preenchido_bfill'] = df_temperaturas['Temperatura'].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'Nome':['Tiago','Gabriel', 'Cristiano', 'Warlley',"Felipe"],
    'Cidade':['Samambaia', np.nan, 'Florianopolis', np.nan, 'Águas Claras']
})

df_cidades['cidade_preenchida'] = df_cidades['Cidade'].fillna('Não Informado')
df_cidades
```

```python
import plotly.express as px
```

```python
df_senioridade_usd = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=False).reset_index()

fig = px.bar(
    df_senioridade_usd,
    x='senioridade',
    y='usd',
    title='Média Salarial por Senioridade',
    labels={'senioridade': 'Senioridade', 'usd': 'Média Salarial Anual (USD)'}
)
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']
fig = px.pie(
    remoto_contagem,
    names = 'tipo_trabalho',
    values = 'quantidade',
    title='Proporção Tipos de Trabalho',
    hole=0.5
)
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']
fig = px.pie(
    remoto_contagem,
    names = 'tipo_trabalho',
    values = 'quantidade',
    title='Proporção Tipos de Trabalho',
    hole=0.5
)
fig.update_traces(textinfo='percent+label')
fig.show()
```

```python
cientista_dados = df_limpo[df_limpo['cargo'].str.contains('Data Scientist', case=True, na=False)]
media_salarial_pais = cientista_dados.groupby('residencia')['usd'].mean().sort_values(ascending=False).reset_index()
fig = px.bar(
    media_salarial_pais,
    x='residencia',
    y='usd',
    title='Média Salarial de Cientistas de Dados por País',
    width = 1400,
    height = 500,
    color_discrete_sequence = px.colors.qualitative.Pastel

)
fig.show()
```

```python
import pycountry

def iso2_to_iso3(code):
  try:
    return pycountry.countries.get(alpha_2=code).alpha_3
  except:
    return None

df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)

df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()

fig = px.choropleth(
    media_ds_pais,
    locations = 'residencia_iso3',
    color ='usd',
    color_continuous_scale='rdylgn',
    title='Salário médio de Cientista de Dados por País',
    labels={'usd': 'Salàrio médio(USD)', 'residencia_iso3': 'País'}
)
fig.show()
```
